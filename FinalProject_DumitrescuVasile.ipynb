Description of Your Project

1.  General description of the project (it could be a project of changes in your life or a project you are working on).
2.  Gantt chart for this project.
3.  Graph with the critical path marked on it.
4.  Conclusions: as a project manager, what areas will you focus on to ensure project deadlines (most critical moments in the project, what needs special attention to prevent delays)?

````python
# Defining Work Tasks and Duration Estimates

tasks = {
    "Overpaying the mortgage": 7,
    "paying off all loans": 1,
    "doubling income": 2,
    "acquiring additional programming skills": 2,
    "investing 70% of income": 8,
    "expanding knowledge of investments": 5,
    "researching local and global investment markets": 5,
    "establishing and developing a consulting firm": 2,
    "writing several industry-specific books": 4,
    "obtaining a master's degree, learning programming": 2,
    "buying farm land": 1,
    }


print(tasks)
````

````python
# Dependencies

dependencies = {
    "paying off all loans": [],
    "Overpaying the mortgage": ["paying off all loans"],
    "doubling income": ["paying off all loans"],
    "acquiring additional programming skills": ["doubling income"],
    "investing 70% of income": ["doubling income"],
    "expanding knowledge of investments": [],
    "researching local and global investment markets": [],
    "establishing and developing a consulting firm": ["writing several industry-specific books"],
    "writing several industry-specific books": ["investing 70% of income"],
    "obtaining a master's degree, learning programming": ["doubling income"],
    "buying farm land": ["doubling income"],
    }
print("Work Sequence:")
for task, deps in dependencies.items():
  print(f"{task}: {deps}")
````

````python
# Form an adjacency matrix in Python.
task_indices = {task: i for i, task in enumerate(tasks)}
adjacency_matrix = np.zeros((len(tasks), len(tasks)), dtype=int)

for task, deps in dependencies.items():
    for dep in deps:
        adjacency_matrix[task_indices[dep], task_indices[task]] = 1

# Build a graph in Python and mark the critical path.
G = nx.DiGraph()

# Dodanie krawędzi z macierzy sąsiedztwa
for i in range(len(adjacency_matrix)):
    for j in range(len(adjacency_matrix)):
        if adjacency_matrix[i][j] == 1:
            G.add_edge(list(tasks.keys())[i], list(tasks.keys())[j])
            print(f"Adding edge from {list(tasks.keys())[i]} to {list(tasks.keys())[j]}")

# Obliczanie czasu najwcześniejszego zakończenia dla każdego zadania
earliest_finish = np.zeros(len(tasks), dtype=int)

def calculate_earliest_finish(task_index):
    predecessors = np.where(adjacency_matrix[:, task_index] == 1)[0]
    if len(predecessors) == 0:
        return tasks[list(tasks.keys())[task_index]]
    else:
        return tasks[list(tasks.keys())[task_index]] + max(calculate_earliest_finish(pred) for pred in predecessors)

for i in range(len(tasks)):
    earliest_finish[i] = calculate_earliest_finish(i)

# Znalezienie ścieżki krytycznej
critical_path = nx.algorithms.dag.dag_longest_path(G)
print(critical_path)
````

Wykres Guntta

````python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# Drawing the Gantt chart
# Creating data points for the Gantt chart
tasks_list = list(tasks.keys())
start_times = earliest_finish
end_times = earliest_finish

# Setting up the figure and axes for the Gantt chart
plt.figure(figsize=(12, 8))
plt.grid(True)

# Drawing bars for each task
for i, task in enumerate(tasks_list):
    predecessors = [list(task_indices.keys())[idx] for idx in np.where(adjacency_matrix[:, i] == 1)[0]]
    if predecessors:
        start_time = max(end_times[task_indices[p]] for p in predecessors)
    else:
        start_time = start_times[i]
    end_time = end_times[i]
    duration = end_time - start_time  # Obliczenie czasu trwania zadania

    # Check if the task is on the critical path
    if task in critical_path:
        plt.barh(y=i, width=end_time - start_time, left=start_time, height=0.4, color="red", align="center")
    else:
        plt.barh(y=i, width=end_time - start_time, left=start_time, height=0.4, color="skyblue", align="center")

    plt.text(start_time + duration / 2, i, f"{duration} years", verticalalignment="center", horizontalalignment="center")

    for predecessor in predecessors:
        plt.arrow(end_times[task_indices[predecessor]], tasks_list.index(predecessor), start_time - end_times[task_indices[predecessor]] - 0.1, i - tasks_list.index(predecessor), head_width=0.2, head_length=0.1, fc='k', ec='k')

# Setting title and axis labels
plt.xlabel("Years")
plt.ylabel("Tasks")
plt.title("Gantt Chart for the Project")

# Changing Y axis labels to match task names
plt.yticks(range(len(tasks_list)), tasks_list, fontsize=8)

# Adding labels to X axis
plt.xticks(range(0, max(earliest_finish) + 1, 2))

# Przygotowanie legendy
red_patch = mpatches.Patch(color='red', label='Critical Path Tasks')
blue_patch = mpatches.Patch(color='skyblue', label='Other Tasks')
plt.legend(handles=[red_patch, blue_patch])

# Displaying the Gantt chart
plt.show()
````

7 years of saving or 7 years of saving & investing

````python
from matplotlib.ticker import StrMethodFormatter

def calculate_simple_savings(months, salary, saving_rate):
    # Simple savings is the sum of a set percentage of one's income saved each month
    return saving_rate * salary * months

def calculate_compound_savings(months, salary, saving_rate, annual_interest_rate):
    # Compound savings calculated with monthly compounding interest.
    compound_savings = np.zeros_like(months, dtype=float)
    monthly_savings = salary * saving_rate
    monthly_interest_rate = (1 + annual_interest_rate)**(1/12) - 1
    for i in range(1, len(months)):
        compound_savings[i] = (compound_savings[i-1] + monthly_savings) * (1 + monthly_interest_rate)
    return compound_savings

# Parameters
salary = 20000  # some sample monthly salary / income
saving_rate = 0.7  # 70% which is going for investment
annual_interest_rate = 0.15  # 15% annual intrest

# Months od 0 do 84
months = np.arange(0, 85)

# Counting savings
simple_savings = calculate_simple_savings(months, salary, saving_rate)
compound_savings = calculate_compound_savings(months, salary, saving_rate, annual_interest_rate)

# Visualising the graph before (saving only)
plt.figure(figsize=(12, 6))
plt.plot(months, simple_savings, label='Just Saving', color='blue')

plt.title('Savings of 70& of Income / Salary For 7 Years')
plt.xlabel('Months')
plt.ylabel('Savings on account')
plt.grid(True)

# Formating Y with ' for thousents
plt.gca().yaxis.set_major_formatter(StrMethodFormatter('{x:,.0f}'))

# Setting rotation of labeling for X
plt.xticks(rotation=45)

# Setting values for Y
max_savings = max(max(simple_savings), max(compound_savings))
plt.yticks(np.arange(0, max_savings, salary*saving_rate*12), rotation=45)  # Saving every year

# Visualising the graph after the change (saving + investing)
plt.figure(figsize=(12, 6))

plt.plot(months, compound_savings, label='Saving + Investing', color='green')
plt.title('Savings + Investing of 70& of Income / Salary For 7 Years')
plt.xlabel('Months')
plt.ylabel('Savings on account')
plt.grid(True)

# Formating Y with ' for thousents
plt.gca().yaxis.set_major_formatter(StrMethodFormatter('{x:,.0f}'))

# Setting rotation of labeling for X
plt.xticks(rotation=45)

# Setting values for Y
max_savings = max(max(simple_savings), max(compound_savings))
plt.yticks(np.arange(0, max_savings, salary*saving_rate*12), rotation=45)  # Saving every year

# Visualising the graph (Both)
plt.figure(figsize=(12, 6))
plt.plot(months, simple_savings, label='Just Saving', color='blue')
plt.plot(months, compound_savings, label='Saving + Investing', color='green')
plt.title('Savings vs Savings + Investing of 70& of Income / Salary For 7 Years')
plt.xlabel('Months')
plt.ylabel('Savings on account')
plt.grid(True)

# Formating Y with ' for thousents
plt.gca().yaxis.set_major_formatter(StrMethodFormatter('{x:,.0f}'))

# Setting rotation of labeling for X
plt.xticks(rotation=45)

# Setting values for Y
max_savings = max(max(simple_savings), max(compound_savings))
plt.yticks(np.arange(0, max_savings, salary*saving_rate*12), rotation=45)  # Saving every year

plt.legend()
plt.tight_layout()
plt.show()
````

Decision Optimization

1.  Description of the situation (optimization of the types of services provided by the company, time of employees on the project, distribution of your own time, etc.).

I inherited a family botanical garden, which has been a place of peace and beauty for the local community for years. However, despite its charm, the garden has financial problems. After conducting an analysis, I realized that some sections of the garden do not generate sufficient income, while others are very popular and could bring greater profits.

I have already identified areas of the garden that can be profitable, but now my task is to optimize the development of the botanical garden under the following conditions:

The garden has three main sections: a section of decorative plants, a vegetable section, and a section of exotic potted plants. Each section requires a specific amount of maintenance work, and costs and revenues depend on the type of plants. For example:

Decorative plants require regular watering, fertilizing, and pruning, but they bring in higher income due to their attractiveness to customers.
The vegetable section requires more work in planting, weeding, and harvesting, but may yield lower income compared to decorative plants.
Exotic potted plants are easier to maintain, but bring in less income because they are less popular.
My task is to optimize the allocation of work and resources in the individual sections of the botanical garden to maximize income.

Decorative plants section:
Monthly maintenance cost: 1000 PLN
Expected monthly income: 3000 PLN
Number of work hours required per month: 80 hours
Vegetable section:
Monthly maintenance cost: 800 PLN
Expected monthly income: 1500 PLN
Number of work hours required per month: 100 hours
Exotic potted plants section:
Monthly maintenance cost: 500 PLN
Expected monthly income: 1000 PLN
Number of work hours required per month: 60 hours
Resource availability:
Number of workers available: 5 people
Each worker can work a maximum of 160 hours per month.

````python
from scipy.optimize import linprog

# Define the coefficients of the objective function
c = [-3000 + 1000, -1500 + 800, -1000 + 500]  # Adjusted objective function with maintenance costs

# Define the coefficients matrix for the inequality constraints (resource availability)
A_ub = [
    [80, 100, 60],  # Work hours for each section
    [-1, 0, 0],     # Maximum number of sections constraint (non-negativity)
    [0, -1, 0],     # Maximum number of sections constraint (non-negativity)
    [0, 0, -1],     # Maximum number of sections constraint (non-negativity)
    [1, 1, 1]       # Maximum number of workers constraint
]

# Define the right-hand side vector for the inequality constraints
b_ub = [160 * 5, 0, 0, 0, 5]

# Solve the linear programming problem
result = linprog(c, A_ub=A_ub, b_ub=b_ub, method='highs')

# Display the results
if result.success:
    print("Optimal distribution of work in the garden:")
    print("Decorative Plants:", int(result.x[0]))
    print("Vegetable Section:", int(result.x[1]))
    print("Exotic Plants:", int(result.x[2]))
    print("Maximum monthly income:", "${:,.2f}".format(-result.fun))
else:
    print("Optimization failed. Please check the constraints.")
````
